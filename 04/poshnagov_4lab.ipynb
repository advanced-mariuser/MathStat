```python
import numpy as np
import pandas as pd
from scipy.stats import poisson
from scipy.stats import kstest
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('women.xlsx')['count']
```

```python
med = df.median()
print(med)
```

```python
n = 1000
bootstrap_median = []
for _ in range(n):
    bootstrap_df = df.sample(len(df), replace=True)
    bootstrap_median.append(bootstrap_df.median() - med)
```

```python
plt.hist(bootstrap_median, color = 'blue', edgecolor = 'black');
```

```python
print(pd.Series(bootstrap_median).std())
```
