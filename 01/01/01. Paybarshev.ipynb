## 1. [# 30] Приведены данные аварий в период с 15 марта 1851 года по 22 марта 1962 года. Был получен набор данных из 190 зарегистрированных временных интервалов (в днях) между последовательными авариями на угольных шахтах, в результате которых погибло десять или более человек. Упорядоченные данные приведены в таблице days between disasters.xlsx.

```python
# Import the libraries
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import seaborn as sns
from KDEpy import FFTKDE
from scipy.stats import norm
```

__1) Выберите ширины ячейки и постройте гистограмму.__

```python
#создание DataFrame (df) с помощью чтения данных из файла
df = pd.read_excel('days between disasters.xlsx')

m = 1 + 3.3 * np.log10(190)
print(m)

plt.hist(df['days'], color = 'blue', edgecolor = 'black', bins = int(2366/m));
```

__2) Выберите полосу пропускания и постройте графики KDE с различными ядрами.__

```python
data = df['days'].values
dataBw = int(1.06 * df['days'].std() * pow(190, -1/5))
print('Полоса пропускания =', dataBw)

x, y = FFTKDE(bw = dataBw, kernel='epa').fit(data)(190) 
plt.plot(x, y, color = 'blue')
x, y = FFTKDE(bw = dataBw, kernel='gaussian').fit(data)(190) 
plt.plot(x, y, color = 'red')
x, y = FFTKDE(bw = dataBw, kernel='triweight').fit(data)(190) 
plt.plot(x, y, color = 'green')

plt.legend(('epa', 'gau', 'triw'))

plt.savefig('days between disasters', dpi = 200, transparent=True, bbox_inches = 'tight')
```

```python
#Коррекция границ
low_bound = 0
data_reflected = np.concatenate((data, 2 * low_bound - data))
                                                                
x_corrected, y_corrected = FFTKDE(bw = dataBw, kernel='epa').fit(data)(190) 
y_corrected[x_corrected <= low_bound] = 0
plt.plot(x_corrected, y_corrected, color = 'blue')
x_corrected, y_corrected = FFTKDE(bw = dataBw, kernel='gaussian').fit(data)(190) 
y_corrected[x_corrected <= low_bound] = 0
plt.plot(x_corrected, y_corrected, color = 'red')
x_corrected, y_corrected = FFTKDE(bw = dataBw, kernel='triweight').fit(data)(190) 
y_corrected[x_corrected <= low_bound] = 0
plt.plot(x_corrected, y_corrected, color = 'green')

plt.legend(('epa', 'gau', 'triw'))

plt.xlim(0, 2500)
plt.savefig('days between disasters', dpi = 200, transparent=True, bbox_inches = 'tight')
```

__3) Постройте график эмпирической функции распределения.__

```python
sns.kdeplot(data, cumulative=True, legend=False, color = 'blue')
plt.title("$F_n(x)$ для данных дней между катастрофами")
plt.ylabel("")
```

__4) Сделайте предварительные выводы.__

Исходя из построенных графиков, можно сделать вывод о том, что катастрофы с гибелью людей на шахтах зачастую происходят довольно редко по вероятности, но если и происходят то зачастую инетрвал между ними минимальный. В среднем количество дней между катастрофами в районе от 0 до 500 дней. И чем больше дней проходит с момента последней катастрофы, тем больше вероятность, что она случится.
